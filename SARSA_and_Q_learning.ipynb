```python
import gymnasium as gym
import numpy as np
import matplotlib.pyplot as plt
```

```python
import numpy as np
import gym

# Set up the environment
env = gym.make('FrozenLake-v1')

# Initialize Q-table with zeros
Q = np.zeros([env.observation_space.n, env.action_space.n])

# Set parameters
num_episodes = 5000
learning_rate = 0.85
gamma = 0.95
epsilon = 1.0
epsilon_decay = 0.999
cumulative_rewards = []
# SARSA
for i in range(num_episodes):
    # Reset the state
    state = env.reset()
    action = np.argmax(Q[state,:] + np.random.randn(1, env.action_space.n) * (1. / (i+1)))
    cumulative_reward = 0
    for j in range(100):
        # Perform action and get new state & reward
        new_state, reward, done, _ = env.step(action)
        cumulative_reward += reward
        # Select new action
        new_action = np.argmax(Q[new_state,:] + np.random.randn(1, env.action_space.n) * (1. / (i+1)))

        # Update Q-Table
        Q[state, action] = Q[state, action] + learning_rate * (reward + gamma * Q[new_state, new_action] - Q[state, action])

        state = new_state
        action = new_action

        if done:
            epsilon *= epsilon_decay
            break
    cumulative_rewards.append(cumulative_reward)

print("Q-Table")
print(Q)

print("Cumulative rewards")
print(cumulative_rewards)
```
